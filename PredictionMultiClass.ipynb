```python
# Function to display images

import imageio
import matplotlib.pyplot as plt

def plotimg(path):


    # Specify the path to your image file
    image_path = path

    # Read the image using imageio and convert it to a NumPy array
    image = imageio.imread(image_path)

    # Display the shape of the NumPy array (height, width, channels for a color image)
    print("Image Shape:", image.shape)

    # Display the image (optional)
    plt.imshow(image)
    plt.show()
```

```python
#Testing on Glioma Tumor images

import numpy as np
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model

class_labels = ['Glioma', 'Meningioma','No tumor','Pituitary' ]
# Load the saved model
model = load_model("Omar_Rabbah_MultiClass")

# Load and preprocess the input image
img = image.load_img('Te-gl_0028.jpg', target_size=(256, 256))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
img_array /= 255.0  # Normalize pixel values to between 0 and 1

# Make predictions
predictions = model.predict(img_array)

# Get the predicted class index
predicted_class_index = np.argmax(predictions)

# Map the class index to the corresponding class label
predicted_class_label = class_labels[predicted_class_index]

print("Prediction: Class {}".format(predicted_class_label))

plotimg('Te-gl_0028.jpg')
```
